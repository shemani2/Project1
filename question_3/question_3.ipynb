```python
#find specific jobs in data frame 
selected_job_titles = ["Data Scientist", "Data Engineer", "Data Analyst", "Machine Learning Engineer", "Analytics Engineer"]
data_science_df = data_science_df[data_science_df["job_title"].isin(selected_job_titles)]
data_science_df
```

```python
#filter for specific salary
data_science_df= data_science_df.loc[data_science_df["salary_currency"] == "USD", :]
data_science_df
```

```python

#filter location
data_science_df= data_science_df.loc[data_science_df["company_location"] == "United States", :]
data_science_df
```

```python
#identify unique years

data_science_df["work_year"].unique()
# unique_years = data_science_df["work_year"].unique()
# print(unique_years)
```

```python
#filter our 2024

clean_data_df = data_science_df.loc[data_science_df["work_year"] != 2024]
clean_data_df
```

```python
# Find the maximum and minimum salary in the clean_data_df DataFrame
max_salary = clean_data_df['salary'].max()
min_salary = clean_data_df['salary'].min()

print(f"Maximum Salary: {max_salary}")
print(f"Minimum Salary: {min_salary}")
```

```python
#number of jobs per year 
data_by_year_df = clean_data_df[['job_title', 'work_year']].groupby('work_year').count()
data_by_year_df

# And plot the number of jobs in each year - shows trending increase by year
plt.style.use('ggplot')
plt.plot(data_by_year_df['job_title'])
plt.xticks(rotation=75)
plt.title('Number of Jobs Over Time')
plt.xlabel('Years')
plt.ylabel('Quantity of Jobs')
plt.show()
plt.savefig('Images/NumberofJobs.png')
```

```python
# Group the data by "work_year" and "job_title" and calculate the average salary
average_salary_by_job_title = clean_data_df.groupby(["work_year", "job_title"])["salary"].mean().unstack()

# Sort the data to identify the trending job titles per year based on salary
trending_jobs = average_salary_by_job_title.T.idxmax()
trending_jobs
```

```python
#create table of top trending job per year
from tabulate import tabulate
table = [[2020, 'Machine Learning Engineer'],  
          [2021, 'Machine Learning Engineer'], 
          [2022, 'Machine Learning Engineer'],
          [2023, 'Machine Learning Engineer']]

#define headers
col_names = ["Work Year", "Top Trending Job"]

#display table
print(tabulate(table, headers=col_names))
```

```python
trending_jobs = clean_data_df.groupby(["work_year", "job_title"])["salary"].mean().unstack().fillna(0)
print(trending_jobs)

trending_jobs.set_index(pd.Index(['2020', '2021', '2022', '2023']))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a box plot to determine 
plt.figure(figsize=(12, 8))
sns.boxplot(x='work_year', y='salary', hue='job_title', data=clean_data_df)
plt.title('Distribution of Salaries per Year for Selected Job Titles')
plt.xlabel('Work Year')
plt.ylabel('Salary Amount (USD)')
plt.xticks(rotation=45)
plt.legend(title='Job Title')
plt.show()
plt.savefig('Images/BoxPlotDistribution.png')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a violin plot
plt.figure(figsize=(15, 10))
sns.displot(x='work_year', y='salary', hue='job_title', data=clean_data_df)
plt.title('Distribution of Salaries per Year for Selected Job Titles')
plt.xlabel('Work Year')
plt.ylabel('Salary Amount (USD)')
plt.xticks(rotation=45)
plt.legend(title='Job Title')
plt.show()
plt.savefig('Images/DisplotJobDistribution')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a violin plot
plt.figure(figsize=(15, 10))
sns.kdeplot(x='work_year', y='salary', hue='job_title', data=clean_data_df)
plt.title('Distribution of Salaries per Year for Selected Job Titles')
plt.xlabel('Work Year')
plt.ylabel('Salary Amount (USD)')
plt.xticks(rotation=45)
plt.legend(title='Job Title')
plt.show()
plt.savefig('Images/KdeplotJobDistribution')
```
